# Task 1
Create a dictionary IP_Count where each key-value pair is IP-Address and its frequency in the entire file.
For example, {“10.128.2.1”: 10, “10.131.2.1”: 12 } indicates the IP 10.128.2.1 has appeared 10 times,  10.131.2.1 appeared 12 times in the file and so on.


```python
import re
fhand = open('weblog.txt')
s = fhand.read()    
```

```python
# Creating regular expression for searching IP address
pattern1 = re.findall('\d{2}.\d{3}.\d.\d',s)
pattern1
```

```python
# Counting no. of occurrences
ip_count = {}
for i in pattern1:
    ip_count[i]=ip_count.get(i,0)+1
ip_count
```

# Task 5
Create a dictionary Status_Code where key is the status code, and value is a tuple consisting of a string stating meaning of the code and a number indicating the frequency of that code in the file.
For example, {200:(“Success”, 53), 404: (“File Not Found”,4)} indicates 200 is the code to indicate ‘Success’, and there are 53 such transactions. 


```python
pattern5 = re.findall('\t(\d{3})\n',s)  # Parenthesis is used to extract a paticular subset in this case only code 200.
pattern5
```

```python
# Counting no. of occurrences
code = {}
for i in pattern5:
    code[i]=code.get(i,0)+1
code
```

```python
# Creating a dictionary consisting of a string stating meaning of the code and 
# a number indicating the frequency of that code in the file
status_code= {}

status_code[200]= 'Success',code['200']
status_code[302]= 'Requested resource has been temporarily moved',code['302']
status_code[304]= 'File Not Modified (means, no update is required)',code['304']
status_code[206]= 'Partial content has been processed',code['206']
status_code[404]= 'File Not Found',code['404']
```

# Task 6
Create a histogram (list of tuples) indicating number of actions carried out in an hour of a day. For example [(8, 12), (15, 3)] indicates, there are 12 transactions between 8am to 8:59am, there are 3 transactions between 3pm to 3:59pm (because 15 means 3pm).

```python
pattern6 = re.findall(':\d{2}:\d{2}:\d{2}',s)
pattern6
```

```python
# Counting no. of occurrences
histogram ={}
for i in pattern6:
    histogram[int(i[1:3])] = histogram.get(int(i[1:3]),0)+1 
    # since we have to count for every hour therefore selecting only the hour and changing it to integer  
histogram
```
